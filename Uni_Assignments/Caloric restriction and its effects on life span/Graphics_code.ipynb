```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_x = np.array([0, 0.2, 0.6, 0.7, 0.8, 0.9, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 2.2])
data_y = np.array([1, 0.9, 0.9, 0.85, 0.85, 0.80, 0.80, 0.78, 0.74, 0.74, 0.65, 0.60, 0.60])
blue_x = np.array([0, 0.1, 0.3, 0.4, 0.5, 0.7, 0.8, 0.9, 1, 1.1, 1.3, 1.5, 1.6, 1.8, 2, 2.1])
blue_y = np.array([1, 1, 0.95, 0.95, 0.90, 0.90, 0.80, 0.76, 0.76, 0.75, 0.73, 0.68, 0.63, 0.48, 0.48, 0.47])
plt.plot(blue_x, blue_y, '-b', label='Control')
plt.plot(data_x, data_y, '-r', label='CR')
plt.legend(loc="lower left")
plt.title('Overall mortality')
plt.xlabel("Age (years)")
plt.ylabel("Survival", rotation = 'vertical')
plt.yticks([0, 0.25, 0.50, 0.75, 1], ['0%','25%', '50%', '75%', '100%'])
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5], [10, 15, 20, 25, 30, 35])
plt.xlim(0, 2.5)
plt.ylim(0, 1.1)


plt.show()
```

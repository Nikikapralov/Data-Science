```python
def get_difference(starting, ending):
    zipped = zip(starting, ending)
    difference = [round((ending - starting), 2) for starting, ending in zipped]
    return difference
```

```python
difference_length = get_difference(list(reversed([4.1, 4.2, 4.1, 4.1, 4.2, 4.1, 4.1])), list(reversed([4.1, 4.1, 4.2, 4.2, 4.3, 4.3, 4.5])))
```

```python
difference_weight = get_difference(list(reversed([7.15, 5.57, 4.82, 5.79, 4.15, 4.28, 4.22])), list(reversed([6.43, 5.15, 4.65, 5.93, 4.49, 4.75, 5.25])))
```

```python
concentrations = list(reversed([0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.0]))
```

```python
print(difference_length)
```

```python
def draw_diagram(x, y, title, x_label, y_label):
    plt.plot(x, y)
    ax = plt.gca()
    ax.spines["bottom"].set_position("zero")
    ax.spines["left"].set_position("zero")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    zipped = zip(x, y)
    for item in zipped:
        plt.plot(item[0], item[1], '-ok')
    plt.show()

    
```

```python
draw_diagram(concentrations, difference_weight, 
             'Molar concentration and weight change', 'Molar Concentration (M)',
             r"$\Delta G$ in mg")
```

```python
draw_diagram(concentrations, difference_length,
             'Molar concentration and length change', 
             'Molar Concentration (M)', r'$\Delta L$ in cm')
```
